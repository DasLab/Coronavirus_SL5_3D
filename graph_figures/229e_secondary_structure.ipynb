```python
from arnie.bpps import bpps
from arnie.utils import convert_dotbracket_to_bp_list
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
```

```python
seq = 'ggaagcaagugcuguguguucuagucuaaggguuucguguuccgucacgagauuccauucuacaaacgccuuacgcgagguucugucucguguuuguguggaagcaaaguuuugucuuuguggaaaccaguaacuguucc'
bpp = bpps(seq,package='eternafold')
man = '((((.(.(((((((.(((((.(((.....((((((((((......))))))))))....)))..)))))...(((((((.......)))))))..........((((((......))))))......)))).))))))))'
man = convert_dotbracket_to_bp_list(man)
```

```python
fig,axs=plt.subplots(4,20,figsize=(120,20))
p='eternafold'
for i in range(0,len(seq)-60,5):
    part = seq[i:i+60] #'a'*i+seq[i:i+40]+'a'*(len(seq)-i-40)
    bpp = bpps(part,package=p)
    axs[0,i//5].imshow(bpp)
    axs[0,i//5].set_xticks(list(range(len(part))))
    axs[0,i//5].set_xticklabels(part,size=4)
for i in range(0,len(seq)-80,5):
    part = seq[i:i+80] #'a'*i+seq[i:i+40]+'a'*(len(seq)-i-40)
    bpp = bpps(part,package=p)
    axs[1,i//5].imshow(bpp)
    axs[1,i//5].set_xticks(list(range(len(part))))
    axs[1,i//5].set_xticklabels(part,size=4)
for i in range(0,len(seq)-40,5):
    part = seq[i:i+40] #'a'*i+seq[i:i+40]+'a'*(len(seq)-i-40)
    bpp = bpps(part,package=p)
    axs[2,i//5].imshow(bpp)
    axs[2,i//5].set_xticks(list(range(len(part))))
    axs[2,i//5].set_xticklabels(part,size=4)
for i in range(12,42,2):
    part = seq[:i]+seq[-i:] #'a'*i+seq[i:i+40]+'a'*(len(seq)-i-40)
    bpp = bpps(part,package=p)
    axs[3,(i-12)//2].imshow(bpp)
    axs[3,(i-12)//2].set_xticks(list(range(len(part))))
    axs[3,(i-12)//2].set_xticklabels(part,size=4)
plt.savefig('229e_bpps.png',dpi=300)
```

```python
def plot_bpp(bpp,axs,boundaries=True):
    axs.imshow(bpp,cmap='Greys')
    axs.set_xticks(list(range(len(seq))))
    axs.set_xticklabels(seq,size=10)
    axs.set_yticks(list(range(len(seq))))
    axs.set_yticklabels(seq,size=10)
    
    alpha = 0.05
    
    i_stem = 24
    i_c = 99
    j_c = 124
    i_b = 50
    j_b = 110
    i_a = 10
    j_a = 70
    
    # draw boundaries
    if boundaries:
        plt.hlines(i_stem,len(seq)-i_stem,len(seq),color='black')
        plt.hlines(len(seq)-i_stem,-0.5,xmax=i_stem,color='black')
        plt.vlines(i_stem,len(seq)-i_stem,len(seq),color='black')
        plt.vlines(len(seq)-i_stem,-0.5,i_stem,color='black')
        plt.hlines(i_c,i_c,j_c,color='red')
        plt.hlines(j_c,i_c,j_c,color='red')
        plt.vlines(i_c,i_c,j_c,color='red')
        plt.vlines(j_c,i_c,j_c,color='red')
        plt.hlines(i_b,i_b,j_b,color='orange')
        plt.hlines(j_b,i_b,j_b,color='orange')
        plt.vlines(i_b,i_b,j_b,color='orange')
        plt.vlines(j_b,i_b,j_b,color='orange')
        plt.hlines(i_a,i_a,j_a,color='blue')
        plt.hlines(j_a,i_a,j_a,color='blue')
        plt.vlines(i_a,i_a,j_a,color='blue')
        plt.vlines(j_a,i_a,j_a,color='blue')
        axs.set_ylim(len(seq)-0.5,-0.5)
        axs.set_xlim(-0.5,len(seq)-0.5)
    
    c0 = 103
    cn = 121
    b0 = 72
    bn = 93
    a0 = 29
    an = 55
    sn = 14
    
    axs.add_patch(Rectangle((-0.5,cn-0.5), len(seq), c0-cn, linewidth=0, alpha=alpha,color='red'))
    axs.add_patch(Rectangle((c0-0.5,-0.5), cn-c0, len(seq), linewidth=0, alpha=alpha,color='red'))
    axs.add_patch(Rectangle((-0.5,bn-0.5), len(seq), b0-bn, linewidth=0, alpha=alpha,color='orange'))
    axs.add_patch(Rectangle((b0-0.5,-0.5), bn-b0, len(seq), linewidth=0, alpha=alpha,color='orange'))
    axs.add_patch(Rectangle((-0.5,an-0.5), len(seq), a0-an, linewidth=0, alpha=alpha,color='blue'))
    axs.add_patch(Rectangle((a0-0.5,-0.5), an-a0, len(seq), linewidth=0, alpha=alpha,color='blue'))
    
    axs.add_patch(Rectangle((-0.5,len(seq)-sn+0.5), len(seq), sn, linewidth=0, alpha=alpha,color='black'))
    axs.add_patch(Rectangle((len(seq)-sn+0.5,-0.5), sn, len(seq), linewidth=0, alpha=alpha,color='black'))
    axs.add_patch(Rectangle((-0.5,-0.5), len(seq), sn, linewidth=0, alpha=alpha,color='black'))
    axs.add_patch(Rectangle((-0.5,len(seq)-0.5), sn, -len(seq), linewidth=0, alpha=alpha,color='black'))
    #axs.add_patch(Rectangle((-0.5,len(seq)-sn), sn, len(seq), linewidth=0, alpha=0.1,color='black'))
    #axs.add_patch(Rectangle((len(seq)-sn,-0.5), len(seq), sn, linewidth=0, alpha=0.1,color='black'))
```

```python
pkgs=['eternafold']
fig,axs=plt.subplots(len(pkgs),figsize=(20,20))

for i,p in enumerate(pkgs): 
    bpp = bpps(seq,package=p)
    i_stem = 24
    i_c = 99
    j_c = 124
    i_b = 50
    j_b = 110
    i_a = 10
    j_a = 70
    bpp_composite = np.zeros_like(bpp)
    bpp_composite[i_c:j_c,i_c:j_c] = np.maximum(bpps(seq[i_c:j_c],package=p),bpp_composite[i_c:j_c,i_c:j_c])
    bpp_composite[i_b:j_b,i_b:j_b] = np.maximum(bpps(seq[i_b:j_b],package=p),bpp_composite[i_b:j_b,i_b:j_b])
    bpp_composite[i_a:j_a,i_a:j_a] = np.maximum(bpps(seq[i_a:j_a],package=p),bpp_composite[i_a:j_a,i_a:j_a])
    stem_p = bpps(seq[:i_stem]+seq[-i_stem:],package=p)
    bpp_composite[:i_stem,-i_stem:] = np.maximum(stem_p[:i_stem,-i_stem:],bpp_composite[:i_stem,-i_stem:])
    bpp_composite[-i_stem:,:i_stem] = np.maximum(stem_p[-i_stem:,:i_stem],bpp_composite[-i_stem:,:i_stem])
    plot_bpp(bpp_composite,axs)
plt.savefig('229e_bpp_composite_eternafold.png',dpi=300)
```
